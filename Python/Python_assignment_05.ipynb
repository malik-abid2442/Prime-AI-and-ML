```python
import json
# create cities.json file with data
data = {'cities': ['Islamabad', 'Lahore', 'Karachi'], 'population':[30000000, 20000000, 15000000]}

with open('cities.json', 'w') as f:
	json.dump(data, f)

# load the cities.json file
cities = json.load(open('cities.json','r'))
for key, value in cities.items():
	print(f"{key}: {value}")
```

```python
city = 'Ccw'
pop = 5000000
data['cities'].append(city)
data['population'].append(pop)
with open('cities.json', 'w') as f:
	json.dump(data, f)
 
# load the cities.json file
cities = json.load(open('cities.json','r'))
for key, value in cities.items():
	print(f"{key}: {value}")
```

```python
def update(city, pop):
    try:
        # data = json.load(open('cities.json','r'))
        if city not in data['cities']:
            data['cities'].append(city)
            data['population'].append(pop)
            with open('cities.json', 'w') as f:
                json.dump(data, f)
        else:
            print(f"{city} already exists in the data.")
        
        for key, value in data.items():
            print(f"{key}: {value}")
    except Exception as e:
        print("An error occurred:", e)
update('Multan', 4000000)
update('Faisalabad', 3500000)
update('Lahore', 22000000)
```
