```python
class BankAccount:
    def __init__(self, account_holder,account_number, balance=0):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance
    def deposit(self,amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")
    def withdraw(self,amount):
        if 0 < amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew: {amount}. New balance: {self.balance}")
        else:
                print("Insufficient funds or invalid withdrawal amount.")
    def get_balance(self):
        print(f"Current balance: {self.balance}")
# Example usage:
account = BankAccount("John Doe", "123456789")
account.deposit(500)
account.withdraw(200)
account.get_balance()
```

```python
class Book:
    def __init__(self,title, author, list_of_reviews=None):
        self.title = title
        self.author = author
        if list_of_reviews is None:
            self.list_of_reviews = []
        else:
            self.list_of_reviews = list_of_reviews
    def add_review(self,review):
        self.list_of_reviews.append(review)
        print(f"Review added: {review}")
    def count_reviews(self):
        print(f"Total reviews: {len(self.list_of_reviews)}")
    def display_reviews(self):
        for i in self.list_of_reviews:
            print(f"- {i}")
# Example usage:
book = Book("The Great Gatsby", "F. Scott Fitzgerald")
book.add_review("A timeless classic!")
book.add_review("Loved the characters and plot.")
book.count_reviews()
book.display_reviews()
        
```

```python
class Student:
    def __init__(self,name,roll_no,marks):
        self.__name = name
        self.__roll_no = roll_no
        self.__marks = marks
    def get_name(self,):
        return self.__name
    def get_roll_no(self):
        return self.__roll_no
    def get_marks(self):
        return self.__marks
    
    def set_name(self,name):
        self.__name = name
    def set_roll_no(self,roll_no):
        if 0<roll_no<100:
            self.__roll_no = roll_no
        else:
            print("Invalid roll number. It must be between 1 and 99.")
    def set_marks(self,marks):
        if 0<=marks<=100:
            self.__marks = marks
        else:
            print("Invalid marks. It must be between 0 and 100.")

# Example usage:
student = Student("Alice", 25, 88)
print(student.get_name())
print(student.get_roll_no())
print(student.get_marks())
        
```

```python
class Shape:
    def area(self):
        pass
class Rectangle(Shape):
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        import math
        return math.pi * (self.radius ** 2)
class Triangle(Shape):
    def __init__(self,base,height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height
# Example usage:
shapes = [
    Rectangle(4, 5),
    Circle(3),
    Triangle(6, 8)]
for shape in shapes:
    print(f"Area: {shape.area()}")
```

```python
class Vehicle:
    brand = "Generic"
    model = "ModelX"

class Car(Vehicle):
    def __init__(self,seats):
        self.seats = seats
    def car_info(self):
        print(f"Car Brand: {self.brand}, Model: {self.model}, Seats: {self.seats}")
class Bike(Vehicle):
    def __init__(self,engine):
        self.type_of_bike = engine
    def bike_info(self):
        print(f"Bike Brand: {self.brand}, Model: {self.model}, Type: {self.type_of_bike}")
# Example usage:
car = Car(5)
car.car_info()
bike = Bike("Sport")
bike.bike_info()
```

```python
from abc import ABC, abstractmethod
class Employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass

class Intern(Employee):
    def __init__(self,hourly_wage,hours_worked):
        self.hourly_wage = hourly_wage
        self.hours_worked = hours_worked
    def calculate_salary(self):
        return self.hourly_wage * self.hours_worked

class FullTimeEmployee(Employee):
    def __init__(self,monthly_salary):
        self.monthly_salary = monthly_salary
    def calculate_salary(self):
        return self.monthly_salary
class ContractEmployee(Employee):
    def __init__(self,project_fee):
        self.project_fee = project_fee
    def calculate_salary(self):
        return self.project_fee
# Example usage:
emp1 = FullTimeEmployee(3000)
print(f"Full-Time Employee Salary: {emp1.calculate_salary()}")
emp2 = ContractEmployee(5000)
print(f"Contract Employee Salary: {emp2.calculate_salary()}")

emp = Intern(15, 120)
print(f"Intern Salary: {emp.calculate_salary()}")
```

```python
class Herbivore:
    def eat_plants(self):
        print("Eating plants.")
class Carnivore:
    def eat_meat(self):
        print("Eating meat.")
class Omnivore(Herbivore, Carnivore):
    def eat_both(self):
        print("Eating both plants and meat.")
class Bear(Omnivore):
    def bear_info(self):
        print("I am a bear, an omnivorous animal.")
        self.eat_plants()
        self.eat_meat()
        self.eat_both()
# Example usage:
bear = Bear()
bear.bear_info()
```
