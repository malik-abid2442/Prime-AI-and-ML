```python
dict = {'ali':60, 'umar':70, 'ahmed':80, 'zain':90}
opt = input("enter key to perform operations ")
if opt.upper()=='A':
    # add a new key/value pair
    key = input("enter key to add ")
    dict[key] = int(input("enter value "))
elif opt.upper()=='B':
    key = input("enter key to update value ")
    if key in dict:
        dict[key] = int(input("enter new value "))
    else:
        print(f"{key} not found to update")
elif opt.upper()=='C':
    ser = input("enter key to search ")
    if ser in dict:
        print(f"{ser} found with value {dict[ser]}")
    else:
        print(f"{ser} not found")
elif opt.upper()=='D':
    for key, value in dict.items():
        print(f"{key}: {value}")
else:
    print("invalid option")
    
print("final dictionary:", dict)
```

```python
l1 = [1, 2, 3, 4, 5]
l2 = [2,4,6,8,10]
l3 = list(set(l1) & set(l2))
print("common elements are:", l3)

l1 = list(set(l1)- set(l3))
l2 = list(set(l2)- set(l3))
print("unique elements in list 1 are:", l1)
print("unique elements in list 2 are:", l2)

l4 = l1 + l2
print("merged unique list is:", l4)
```

```python
lis = [1, 3, 5, 8, 10, 6,2,2,2,3,3,3]
lis2 = []
for i in lis:
    if lis.count(i)>1:
        if lis.count(i) not in lis2:
            print(f"{i} occurs {lis.count(i)} times")
            lis2.append(lis.count(i))
print("list of counts of duplicate elements:", lis2)
```

```python
characters = 'aabcdeff ghijjja'
unique = set(characters)
print("unique characters are:", unique)
```

```python
unq_list = []
count = 0
for i in characters:
    if i not in unq_list:
        unq_list.append(i)
        count +=1
print("unique characters list is:", unq_list)
print("number of unique characters are:", count)
    
```
