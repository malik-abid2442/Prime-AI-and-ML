### Python Slicing

```python
word = 'python'
print(word[::-1])
```

```python
print(word[1:4])
```

```python
print(word[-5:-2])
```

```python
word='Prime'
print(len(word))
```

```python
l1 = [1, 2, 3, 4, 5]
l2 = ['a', 'b', 'c', 'd', 'e']
for i in l1:
    print(i)
```

```python
print(l2)
```

```python
print(l2[2:6:2])
```

```python
l2.append(10)
print(l2)
```

```python
l2.insert(5,12)
print(l2)
```

```python
l1.reverse()
print(l1)
```

```python
dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
print(dict.keys())
```

```python
dict.update({'subject': 'Math'})
print(dict.values())
```

```python
for key, value in dict.items():
    # print(f"{key}: {value}")
    print(value)
```
