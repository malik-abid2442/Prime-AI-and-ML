### Conditional Statements

```python
number = 93

if number>= 84:
    if number>= 90:
        print('A+ Grade')
    else:
        print('A Grade')
elif number>= 70 and number<84:
    print('B Grade')
elif number>= 60 and number<70:
    print('C Grade')
elif number>= 50 and number<60:
    print('D Grade')
else:
    print('Fail')
```

```python
num = 80
status = "Pass" if num >= 50 else "Fail"
print(status)
```

```python
day = "friday"

match day:
    case "monday":
        print("It's Monday!")
    case "tuesday":
        print("It's Tuesday!")
    case "wednesday":
        print("It's Wednesday!")
    case "thursday":
        print("It's Thursday!")
    case "friday":
        print("It's Friday!")
    case "saturday":
        print("It's Saturday!")
    case "sunday":
        print("It's Sunday!")
    case _:
        print("Invalid day!")
```

```python
num = 10
while True:
    num += 1
    if num % 7 == 0:
        print(f"The first number greater than or equal to 10 that is divisible by 7 is: {num}")
        break
    
print(num)
```

```python
num = 0
while num < 10:
    num += 1
    if num % 7 == 0:
        print(f"The first number greater than or equal to 10 that is divisible by 7 is: {num}")
        continue
    
print(num)
```

```python

def fact(num):
    n = 1
    for i in range(1,num+1):
        n = n*i
    return n
print(fact(6))
```

```python
def num(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c
print(num(10, 25, 15))
```
