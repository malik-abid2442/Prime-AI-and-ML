```python
class clsmthd:
    word = "Hello, World!"
    @classmethod
    def class_method(cls,new_word):
        print("This is a class method.")
        cls.word = new_word
print(clsmthd.class_method("Hi there!"))
print(clsmthd.word)
```

```python
class math:
    @staticmethod
    def add(a, b):
        return a + b
print(math.add(5, 7))
```

### Encapsulation

```python
class student:
    def __init__(self,name,age):
        self.name = name
        self.__age = age
    def set_age(self, age):
        self.__age = age
    def get_age(self):
        return self.__age
    def display(self):
        print(f"Name: {self.name}, Age: {self.__age}")
        
stud = student("Bob", 22)
stud.set_age(23)
print(stud.get_age())
stud.display()
# print(stud._age)  # Accessing protected member
print(stud._student__age)  # Accessing private member using name mangling
        
```

### Inheritance
`to use or invoke the constructor of parent class always use super() method`

```python
class college:
    def __init__(self,college_name,fee):
        self.college_name = college_name
        self.fee = fee
class student(college):
    def __init__(self,name,college_name,fee):
        super().__init__(college_name,fee)
        self.name = name
    def display(self):
        print(f"Name: {self.name}, College: {self.college_name}, Fee: {self.fee}")
stud = student("Charlie", "XYZ College", 50000)
stud.display()
```

```python
class office:
    def __init__(self,start_time,end_time):
        self.start_time = start_time
        self.end_time = end_time

class employee(office):
    def __init__(self,name,start_time,end_time):
        super().__init__(start_time,end_time)
        self.name = name
    def display(self):
        print(f"Name: {self.name}, Start Time: {self.start_time}, End Time: {self.end_time}")
emp = employee("David", "9 AM", "5 PM")
emp.display()
```

### Abstraction 

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Bark"
dog = Dog()
print(dog.sound())
```

```python
class animal:
    def sound(self):
        return "Some sound"
class cat(animal):
    def sound(self):
        return "Meow"
cat1 = cat()
print(cat1.sound())
```
