```python
# Write a function to return the count the number of digits in a number n
def number_count_check(num):
    count = 0
    while num > 0:
        num = num // 10
        count += 1
    return count
print(number_count_check(110))
```

```python
def func4(num):
    sum = 0
    while num > 0:
        digit = num % 10
        sum += digit
        num = num // 10
    return sum
print(func4(110))
```

```python
number = int(input("for number check enter 1 and to quit enter 2: "))
while True:
    if number==1:
        num = int(input("Enter another number: "))
        
        if num>=1:
            print("Positive number")
        elif num<0:
            print("Negative number")
        else:
            print("Zero")
    elif number==2:
        print("Goodbye!")
        break
    else:
        print("Invalid input")
        continue
```

```python
def cal(a,b,op):
    if op=='+':
        return a+b
    elif op=='-':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
        return a/b
    else:
        return "Invalid operation"
    
print(cal(10,5,'+'))
```
