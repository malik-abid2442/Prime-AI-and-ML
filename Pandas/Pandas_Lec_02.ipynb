```python
df = sns.load_dataset('tips')
df
```

```python
melted_df = df.melt(
    id_vars = ['sex','smoker','day','time','size'],
    value_vars = ['total_bill','tip'],
    var_name='total_Bill_ammount',
    value_name='value'
    
)
```

```python
melted_df
```

```python
orignal = melted_df.pivot_table(
    index = ['sex','smoker','day','time','size'],
    columns = 'total_Bill_ammount',
    values = 'value',
    aggfunc='mean'  # or use another function as appropriate
)
```
