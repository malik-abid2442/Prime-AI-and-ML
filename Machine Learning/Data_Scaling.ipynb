# ***Data Scaling Technique***

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.scatterplot(data=df)
```

```python
standardscaler= StandardScaler()
scaled_data= standardscaler.fit_transform(df.iloc[:,:-1])
scaled_df= pd.DataFrame(scaled_data, columns=df.columns[:-1])

sns.scatterplot(data=scaled_df)
```

```python
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df.iloc[:,:-1])
print("\nMinMaxScaler output:\n", minmax_scaled)

sns.scatterplot(data=pd.DataFrame(minmax_scaled, columns=df.columns[:-1]))
```
