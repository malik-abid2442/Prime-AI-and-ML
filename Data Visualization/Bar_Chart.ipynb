# ***Bar Plots***

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
plt.bar(df['pclass'],df['fare'])
plt.xlabel('pclass')
plt.ylabel('fare')
plt.show()
```

```python
width = 0.4
plt.bar(df['pclass']-width/2,df['fare'])
plt.bar(df['class']+width/2,df['fare'])
plt.show()
```

```python
import numpy as np

grouped = df.groupby(['pclass', 'sex'])['fare'].mean().unstack()

x = np.arange(len(grouped.index))
width = 0.4
```

```python
plt.bar(x - width/2, grouped['male'], width, label='Male')
plt.bar(x + width/2, grouped['female'], width, label='Female')

plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.xticks(x, grouped.index)
plt.legend()
plt.show()
```
