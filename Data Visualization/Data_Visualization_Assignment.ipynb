# ***Data Vosualization Assignment of Matplotlib And Seaborn***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.normal(loc=70, scale=10, size=200)
```

```python
plt.hist(np.random.normal(loc=70, scale=10, size=200), bins=30)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
mean = 70
sns.histplot(np.random.normal(loc=70, scale=10, size=200), bins=30, kde=True)
plt.axvline(x=mean, color='r', linestyle='--', label='Mean (70)')
plt.title("Histogram with KDE of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,hue="sex",showmeans=True,meanline = True,)
```

```python
penguins = sns.load_dataset("penguins")
```

```python
# plt.plot(penguins['flipper_length_mm'], penguins['body_mass_g'], 'o')
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species')
plt.title("Flipper Length vs Body Mass of Penguins")
plt.xlabel("Flipper Length (mm)")
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', marker='x')
```

```python
pivot =pd.pivot_table(
    data=tips,
    values='tip',
    index='day',
    columns='time',
    aggfunc='sum'   # ya 'count', 'max', 'min'
)
sns.heatmap(pivot, annot=True)
```
