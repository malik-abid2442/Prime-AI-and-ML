# ***Scatter Plots***

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
color = [
    "red" if x < 30 else
    "blue" if 30 < x < 50 else
    "green"
    for x in df['fare']
]
plt.scatter(df['age'],df['fare'],color = color)
plt.colorbar(label="BP")
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(df['age'],df['fare'],label='titanic')
plt.scatter(tips['tip'],tips['total_bill'],label='tips')
plt.legend()
plt.show()
```
