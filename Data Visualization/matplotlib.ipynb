# Matplotlib Remaining Plots

## ***Histogram***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
plt.hist(df['age'].dropna(), bins=30, edgecolor='black')
```

```python
sns.histplot(df['fare'], bins=30, edgecolor='black', kde=True)
```

```python
plt.hist(df['age'].dropna(), bins=30, alpha=0.5, edgecolor='black')
plt.hist(df['fare'].dropna(), bins=30, alpha=0.5,edgecolor='red')
```

```python
plt.hist(df['age'].dropna(), bins=30, edgecolor='black')
plt.axvline(x=df['age'].mean(), color='red', linestyle='--', linewidth=2, label='Mean Age')
```

## ***Box Plot***

```python
# Group the data by 'class' and prepare lists for boxplot
grouped = df.dropna(subset=['age']).groupby('class')['age']
data = [group.values for name, group in grouped]
labels = list(grouped.groups.keys())

plt.boxplot(data, labels=labels)
plt.xlabel('Class')
plt.ylabel('Age')
plt.title('Boxplot of Age by Class')
```

```python
plt.boxplot(df['age'].dropna(), labels=['Age'],showmeans=True,meanline=True)
```

```python
# Filter the dataframe to drop rows where 'age' is NaN, ensuring all arrays have the same length
df_filtered = df.dropna(subset=['age'])

plt.stackplot(df_filtered['pclass'], df_filtered['fare'], df_filtered['age'], colors=['skyblue','lightgreen'], labels=['Age','Fare'])
```
