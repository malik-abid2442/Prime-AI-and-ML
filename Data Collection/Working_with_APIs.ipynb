# ***Working With API's***

```python
import requests
import pandas as pd
```

```python
url = 'https://api.artic.edu/api/v1/artworks'

res = requests.get(url)
```

```python
json_data = res.json()
```

```python
df = pd.json_normalize(json_data['data'])
```

```python
df = df[['id','api_model','api_link','title','main_reference_number']]
```
