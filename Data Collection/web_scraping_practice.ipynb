# ***Scraping Amazon Product Information***

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.amazon.com/s?k=mobile+phones&i=todays-deals&crid=1DH3721DJOQSD&sprefix=mobile+phone%2Ctodays-deals%2C315&ref=nb_sb_noss_2"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3",
    "Accept-Language": "en-US,en;q=0.9",}
```

```python
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.content, 'lxml')
```

```python
res.status_code
```

```python
links = soup.find_all('a', class_='a-link-normal s-line-clamp-2 s-line-clamp-3-for-col-12 s-link-style a-text-normal')
links
```

```python
link = []
for i in links:
    link.append(i.get('href'))
```

```python
for i in link:
    print(i)
```

```python
data = {}
for i in link:
    index = link.index(i) + 1
    url = "https://www.amazon.com" + i
    
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.content, 'lxml')
    
    title_elem = soup.find('span', id='productTitle')
    title = title_elem.get_text().strip() if title_elem else "Title not found"
    
    rating_elem = soup.find('span', id='acrCustomerReviewText')
    rating = rating_elem.get_text().strip() if rating_elem else "Rating not found"
    
    price_elem = soup.find('span', class_='a-offscreen')
    price = price_elem.get_text().strip() if price_elem else "Price not found"
    
    data[index] = {
        'link': url,
        'title': title,
        'Rating': rating,
        'Price': price}
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
df
```

```python
df['Rating'] = df['Rating'].str.replace('(', '', regex=False)
```

```python
df['Price'] = df['Price'].str.replace('PKR', '', regex=False)
```

```python
df['Price'] = df['Price'].str.replace(',', '', regex=False)
```
